### <span style="color:orange">**Relational Operators**</span>

Relational operators in python are : `>`, `<`, `>=`, `<=`, `==`, `!=`

````python
a = 10
b = 20

print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a == b)
print(a != b)
````

### <span style="color:orange">**Logical Operators**</span>

Logical operators are used to combine multiple booleans into one.

Most common logical operators in python are : `and`, `or`, `not`

In case of `and`, If both of the booleans are `True` then only it will give `True`

In case of `or`, If atleast one of the booleans is `True` then it will give `True`

`or` will always give opposite to the given boolean

| A | B | A and B | A or B | not A | not B |
| --- | --- | --- | --- | --- | --- |
| True | True | True | True | False | False |
| True | False | False | True | False | True |
| False | True | False | True | True | False |
| False | False | False | False | True | True 

````python
a = 10
b = 20

c1 = a > 10
c2 = b > 10

r1 = c1 and c2
r2 = c1 or c2
r3 = not c1

print(r1)
print(r2)
print(r3)
````

### <span style="color:orange">**Else-If / Elif Statement**</span>

### <span style="color:orange">Problem :</span> Get highest of 3 numbers

````python
a = 3
b = 3
c = 1

if a >= b and a >= c:
    print(a)
elif b >= a and b >= c:
    print(b)
else:
    print(c)
````

* Order of `if` and `elif` matters a lot.

````python
n = 30

if n > 10:
    print('red')
elif n >= 5:
    print('green')
elif n > 0:
    print('yellow')
````

````python
n = 20

if n >= 5:
    print('green')
elif n > 10:
    print('red')
elif n > 0:
    print('yellow')
````

* Here `red` is correct according to the logic. But the output is `green` just because order matters. When it came to first `if` block and if it matched the condition then it has been executed.

### <span style="color:orange">**Nested Conditional Statements**</span>

````python
n = 2
m = 1

if n % 2 == 0:
    if m % 2 == 0 :
        print("1")
    else:
        print("2")
else:
    print("3")
````

### <span style="color:orange">**While Loop**</span>

`While` loop will figure out, when will the condition fail and stop executing.

```python
while True:
    print(1)
```

* It will print `1` infinite times.

```python
while False:
    print(1)
```

* It will never print `1`

````python
n = 2
count = 1

while count <= n:
    print(1)
    count += 1
````

### <span style="color:orange">Problem :</span> Primality Check

If the number `n` is not divisible by any of the numbers in range `2` to `n-1`, It is called as prime number.

````python
n = 5

div = 2
flag = False
while div < n:
    if n % div == 0:
        flag = True
    div += 1

if flag:
    print('Not a prime number')
else:
    print('prime number')
````

### <span style="color:orange">**Nested While Loop**</span>

### <span style="color:orange">Problem :</span> Print all prime numbers between 2 to n

````python
n = 10

number = 2
while number <= n:
    div = 2
    flag = False
    while div < number:
        if number % div == 0:
            flag = True
        div += 1
    if not flag:
        print(number)
    number += 1
````

### <span style="color:orange">Problem :</span> Reverse of a number

* Last digit of any integer can be found by remainder `n % 10`

* Number excluding last digit can be found by divider `n / 10`

Let's say the number `n = 152`, So reverse of this number will be `251`

The process of reversing a number is as follows :



Let's split `152` into `15` and `2` and update reverse with `(reverse * 10) + remainder` means `(0 * 10) + 2 = 2`



Further split `15` into `1` and `5` and update reverse with `(reverse * 10) + remainder` means `(2 * 10) + 5 = 25`



Further split `1` into `0` and `1` and update reverse with `(reverse * 10) + remainder` means `(25 * 10) + 1 = 251`



We need to repeat the split until we reach `0` as most significant digit.

````python
n = 15463

reverse = 0
while n != 0:
    ms = n // 10
    ls = n % 10
    n = ms
    reverse = (reverse * 10) + ls

print(reverse)
````

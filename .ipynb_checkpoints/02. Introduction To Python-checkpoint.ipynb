### <span style="color:orange">**Data Types in Python**</span>

In python, It creates some storage first then, a varible will get assigned to that storage.

Java gives storage of fixed size with respect to the assigned data type we are storing.

Let's say we assigned a variable `a = 10`. The we changed the value of it by assigning `a = 20`. Then Python will create different storage instead of writing 20 in the same storage like java.

In python, variable `a` is storing some address. Now if `a = 'suryakant'` then python will create another storage and assign the address to the variable.

In java, it might not be able to store big amount of data in the storage assigned previously. In this case, I am restricting myself to what i can store in this storage.

```python
a = 10
a = 20
a = 'suryakant'

a
```

```python
a = 10

type(a)
```

```python
a = 'suryakant'
print(type(a))

a = 10
print(type(a))
```

### <span style="color:orange">**Python Numbers**</span>

```python
a1 = 25
a2 = 1.5
a3 = 5 + 10j

print(type(a1))
print(type(a2))
print(type(a3))
```

```python
a = 10
a = a + 1

a
```

`type()` is used to tell the data type of the given data.

`id()` is used to tell the address of given variable.

```python
a = 10
print(id(a))

a = a + 1
print(id(a))
```

```python
a = 10
print(id(a))

b = 10
print(id(b))
```

```python
a = 1000
print(id(a))

b = 1000
print(id(b))
```

Python only does this optimization of giving same address for same numbers for the range `-5 to 256`. But, Optmizations doesn't work for numbers outside this range.

### <span style="color:orange">**Limit of Integers**</span>

The size of a number determines how much storage it needs, and it can't be larger than the memory available in our system.

In Python, the size of integers isn't limited by the number length but by the memory available for the Python program.

So, the maximum length of an integer in Python is constrained by the memory allowed for the program rather than the inherent length of the number itself.

### <span style="color:orange">**Arithmetic Operators**</span>

Arithmetic operators in python are :

`+`    Addition

`-`    Subtraction

`*`    Multiplication

`/`    Division

`//`   Floor or Integer Division

`**`   Exponentiation

`%`    Modulus

```python
2 + 3 * 4
```

```python
2 * 3 / 4
```

```python
2 * 3 // 4
```

Here `Floor division` operation is done after multiplication. If we want to do floor division first then, we'll have to use paranthesis `()`

```python
2 * (3 // 4)
```

```python
# Simple interest to lowest integer

p = 100
r = 12.5
t = 3
si = (p * r * t) // 100

si
```

```python
# Fahrenheit to Celcius Conversion

f = 100
c = (f - 32) * 5 / 9

c
```

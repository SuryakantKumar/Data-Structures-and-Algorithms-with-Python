### <span style="color:orange">**What is Flow chart ?**</span>

A flow chart is a Diagrammatic representation that shows the steps to solve a problem. It breaks the solution into smaller steps and makes it easier to understand.

Here are a few good things about using flow charts :

* **Debugging :** You can find and fix small mistakes by testing them in the flow chart.

* **Language Independent :** Even if one person codes in Python and another in Java, they can both understand the solution in the flow chart. It helps people from different programming backgrounds understand each other's ideas.

* **Simplification :** It breaks down complicated stuff into smaller, easier steps. This makes it simpler for everyone to understand.

* **Documentation :** It's like taking a picture of the solution to the problem, so you can always look back and remember how you solved it.

### <span style="color:orange">**Decision Making Algorithms**</span>

Algorithms use conditions to decide what to do. They check if something is `true` or `false` and then take different actions based on that.

If something is `true`, the algorithm does one thing. If it's `false`, it does something else. This helps algorithms make choices and do the right thing depending on the situation.

Conditions are shown by rhombus box in the flow chart.
